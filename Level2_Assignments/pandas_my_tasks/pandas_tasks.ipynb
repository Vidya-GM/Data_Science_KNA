## Task: 
**In the dataframe below:**
1. Select only user_id and revenue.

2. Filter users from US.

3. Filter users from US with revenue > 400.

4. Set revenue to 0 for country == "FR".

```python
df = pd.DataFrame({
    "user_id": [1, 2, 3, 4],
    "country": ["US", "IN", "US", "FR"],
    "revenue": [500, 2000, 300, 700]
})
```

```python
# 1. Select only user_id and revenue.
df["user_id"]
```

```python
# 2. Filter users from US
df[df["country"]== "US"]["user_id"]
df.loc[(df["country"]== "US"), "user_id"]
```

```python
# 3. Filter users from US with revenue > 400.
df.loc[((df["country"] == "US") & (df["revenue"]>400)), "user_id"]
```

```python
# 4. Set revenue to 0 for country == "FR".
df.loc[df["country"] == "FR", "revenue"] = 0
df
```
