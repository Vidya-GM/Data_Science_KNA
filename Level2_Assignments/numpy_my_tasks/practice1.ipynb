#### Boolean Indexing Returns a Copy, Not a View
    Its example is shown below
    

```python
# ------------------

a = np.array([1, 2, 3, 4, 5, 6, 7])
b = a[1:4]
b[0] = 99
print(a)

#------------------

```

```python
# Boolean Indexing Returns a Copy, Not a View

arr1 = np.array([5, 10, 15, 20, 25, 30])
sub_arr1 = arr1[arr1>20] # Gives a copy not view
sub_arr1[::] = 0
print(arr1[arr1>20])
print(arr1) # Therefore array is not changed!
```

```python
arr = np.array([5, 10, 15, 20, 25, 30])
print(arr[-3:])
print(arr[::2])
print(arr[arr>20]) 
# Here we are printing a slice-(in terms of condition/boolean-mask, ie only that part where condition is True) of an array
arr[arr>20] = 0
# Here we are modifying arr by providing mask-in place ie without creating its copy
print(arr)
```

#### Broadcasting (ðŸ”¥ VERY IMPORTANT ðŸ”¥)

**What is Broadcasting?**
- Broadcasting allows NumPy to perform operations on arrays of **different shapes**.

```python
# 1. Example 1: Scalar + Array

arr = np.array([10, 20, 30])
print(arr + 5)
```

```python
# Example 2: Row-wise operation
data = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(data + np.array([1, 2, 3]))
```

```python
data1 = np.array([
                [1, 2, 3],
                [4, 5, 6],
                [2, 2, 2]
                ])
data2 = np.array([1, 2, 3])  # Column-wise --> Row remains constant say row1(data1) and columns of this row will be added to data2 matrx
print(data2.shape)
data3 = np.array([[1], [2], [3]]) # Row-wise --> Column remains constant. data 3 has one column. So column of data1 remains constant say column1, 
# and this one column of data3 will be added to all columns of data1
print(data3.shape)

#print(data1+data2.T)
print("-----------------")
print(data1+data2)
print(data1+data3)
```

#### Broadcasting Practice

**If the added array has shape (1, n) â†’ applied to each row** ---> column-wise

**If the added array has shape (n, 1) â†’ applied to each column** ----> row-wise

```python
a = np.ones((3,3))
x = np.array([1, 2, 3]) 
print(a+x)
# 2,3,4
# 2,3,4
# 2,3,4
print(a+x.T)
print(a+x.reshape(3,1))
# 2, 2, 2
# 3, 3, 3
# 4, 4, 4
```

```python
# 5ï¸âƒ£ Real-World Example: Feature Scaling

data = np.array([
    [25, 50000],
    [30, 60000],
    [35, 70000]
], dtype=float)

means = data.mean(axis=0)
stds = data.std(axis=0)
print(means, stds)

normalised = (data-means)/stds
print(normalised)
```

```python
# Practice problem

data = np.array([
    [160, 55],   # height (cm), weight (kg)
    [170, 65],
    [180, 75],
    [190, 85]
], dtype=float)

# Compute the mean of each column
mean_of_data = data.mean(axis=0)   # Calculates mean per column
# Compute stds of each column
stds_of_data = data.std(axis=0)    # Calculates std per column
print(mean_of_data, stds_of_data)
# Compute normalised data
normalised_data = (data-mean_of_data)/stds_of_data # mean, and std are matrix which interact with data elementwise for each row
print(normalised_data)

# Check
print(normalised_data.mean(axis=0))
print(normalised_data.std(axis=0))
```
